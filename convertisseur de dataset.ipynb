```python
import pandas as pd
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import KNNImputer,SimpleImputer,IterativeImputer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
import random
```

```python
data = {
    'student_id': [f'Student_{i+1}' for i in range(250)],
    'gender': np.random.choice(['Male', 'Female', np.nan], size=250),
    'age': np.random.randint(18, 25, size=250),
    'course': np.random.choice(['Math', 'History', 'Science', np.nan], size=250),
    'grade': np.random.choice(['A', 'B', 'C', np.nan], size=250),
    'attendance_percentage': np.random.uniform(70, 100, size=250),
    'extracurricular_activity': np.random.choice(['Yes', 'No', np.nan], size=250),
    'pass_exam': np.random.choice([1, 0], size=250)
}


df = pd.DataFrame(data)
df.replace('nan', np.nan, inplace=True)
df.tail(100)
```

```python
df.isnull().sum()
```

```python



# Ajouter certaines lignes complètement vides
for _ in range(50):  # Ajouter 5 lignes vides
    df.loc[df.shape[0]] = [np.nan] * df.shape[1]

# Réorganiser les indices du DataFrame
df.reset_index(drop=True, inplace=True)
df = df.sample(frac=1).reset_index(drop=True)

# Afficher le DataFrame
df.tail(10)
```

```python
data_manquante = df

# # Enregistrer le nouveau dataset avec des valeurs manquantes
nom_fichier_manquant = "df_academic_cleaned.csv"
df.to_csv(nom_fichier_manquant, index=False)
```
